# Transformations of variables

we measura a variable x, but the interesting final result is y(x). If we know the PDF p(x), what is the PDF p(y)?

$p(y) = p[\phi^{-1}(y)] \abs(\partial_y \phi^{-1}(y))$

where y = phi(x).

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math 
from numpy import random
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
random.seed(42)
n = 10000
x = random.rand(n)

n_bars = 60 #numero de barras de nuestro histograma

x = np.float32(x)
    
plt.plot(x,'.', color = 'lightpink')
plt.show()

plt.hist(x,n_bars,color = 'palevioletred', edgecolor = 'mediumvioletred');
```

```python
#Con esta información obtendremos el punto medio de cada barra del histograma, para una gráfica bonita
x_hist_counting = np.histogram(x,n_bars)[0]
x_hist_range = np.histogram(x,n_bars)[1]
t = np.zeros(len(x_hist_counting)) #nuevo eje x

y = np.zeros(len(x_hist_counting)) #Nueva variable que tendrá la infomración de la func. de x
y = np.float64(y)


#Aquí encontramos el punto medio de cada caja del histograma
for i in range(1,len(x_hist_range)):
    t[i-1]=(x_hist_range[i] + x_hist_range[i-1])/2
    
#    aqui introducimos la forma de los datos de x
for i in range(len(x_hist_counting)):
    y[i]=math.exp(t[i])
    
plt.plot(t,y, color = 'palevioletred')
```
